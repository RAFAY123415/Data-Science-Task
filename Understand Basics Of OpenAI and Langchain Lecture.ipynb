#Model:
Specifies the version of the AI to use. "gpt-3.5-turbo" is used here, but you can choose other versions based on your needs.

#Prompt:
The text or question you provide for the AI to respond to. In this case, it’s "Explain the theory of relativity."

Limits the response length to 150 tokens. This ensures the output is concise and within a specified length.

#Temperature:
 Set to 0.7 to balance creativity and coherence. A lower temperature makes responses more focused, while a higher one makes them more varied.

#Top-p:
 Set to 0.9 to limit the tokens considered to those that make up the top 90% of the probability mass. This helps control the diversity of the response.

#Frequency Penalty:
 Set to 0.5 to reduce repetition in the generated text. This helps in avoiding repeated phrases or words.

#Presence Penalty:
 Set to 0.7 to encourage the model to introduce new topics and avoid sticking to previously mentioned ones.

#Stop:
 The response generation will stop at a newline character ("\n"), ensuring the output ends neatly.

#n:
 Set to 3 to request 3 different responses to the same prompt. This gives you multiple options to choose from.

#Logprobs:
 Shows the log probabilities of the top 5 tokens, useful for understanding the model’s decision-making process.

#Echo:
Set to False, meaning the prompt won’t be included in the output. If True, the prompt would be echoed along with the response.

#Stream:
 Set to False, meaning the response is not streamed in real-time. If True, you’d receive the response as a stream of tokens, useful for real-time applications.

#Explanation of Key Parameters:

##functions:

This is a list of functions that you define and provide to the model. The model can call these functions when it deems appropriate based on the conversation.
In your case, student_custom_function is the function being passed to handle tasks like processing student data.

##function_call='auto':

When you set function_call='auto', you're telling the model to automatically decide whether to invoke a function based on the content of the conversation.

If the model detects that it needs to use the provided function (based on the user's request or the structure of the input), it will "call" that function and return a result.

If no function is relevant, it will continue generating a regular response instead of triggering the function.

```python
import json
student_info = [student_description_two,student_description_three]
for student in student_info:
    response =  client.chat.completions.create(
        model = 'gpt-3.5-turbo',
        messages = [{'role': 'user', 'content': student}],
        functions = student_custom_function, # The Function Is Case Sensative
        function_call = 'auto'
    )

    response = json.loads(response.choices[0].message.function_call.arguments)
    print(response)#import csv
```

#Templates

In LangChain, templates are primarily used in prompt engineering to structure and customize the input provided to large language models. There are a few key types of templates in LangChain:

#StringPromptTemplate:

A basic template that allows you to define static strings with variables to be dynamically substituted when generating a prompt. You can use string formatting or more advanced methods to replace variables within the template.

#ChatPromptTemplate:

A specialized template designed for chat-based interactions (like those with OpenAI's ChatGPT). It structures conversations with a sequence of messages (e.g., from human, AI, or system roles).

#FewShotPromptTemplate:

A type of template used to provide a few examples (few-shot learning) along with the input prompt. This can help the model understand the task better by showing what the input and output should look like before the actual task is performed.

#OutputParser:

Although not strictly a "template," output parsers are often used in conjunction with templates to process and format the model's responses into structured formats (e.g., JSON, tables, etc.).

# String Prompt Template

#Example 1: Using StringPromptTemplate

You can create a string template with placeholders and later substitute them with actual values. Here's an example:

```python
from langchain.prompts import PromptTemplate

template = "Translate the following phrase to French: {phrase}"
prompt = PromptTemplate(template=template, input_variables=["phrase"])

# Filling the prompt with a specific phrase
formatted_prompt = prompt.format(phrase="How are you?")
print(formatted_prompt)  # Output: Translate the following phrase to French: How are you?
```

#Example 2: Using StringPromptTemplate


```python
from langchain.prompts import PromptTemplate

# Create a string prompt template
template = """
You are a helpful assistant. Here is some context:
{context}

Now, please answer the following question:
{question}
"""

# Create a PromptTemplate object
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=template,
)

# Fill the template with actual values
filled_prompt = prompt.format(
    context="The topic is artificial intelligence and its applications.",
    question="What are the benefits of AI in healthcare?"
)

print(filled_prompt)
```

#Example 2: Using ChatPromptTemplate

The ChatPromptTemplate is used when you're working with a chat-based model like OpenAI's GPT-4 or ChatGPT. It structures the interaction by creating messages for different roles (like Human, AI, or System).

```python
from langchain.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# Define a system message and a human message
system_template = "You are a knowledgeable assistant that answers questions based on provided information."
system_message_prompt = SystemMessagePromptTemplate.from_template(system_template)

human_template = "Here is a question: {question}"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

# Create a chat prompt template
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

# Fill the template with actual values
messages = chat_prompt.format_messages(question="What is the importance of data privacy in AI?")

# Print the content of the messages
for message in messages:
    print(message.content)
```

#Example: Few-Shot Prompt Template in LangChain

Suppose you are asking the language model to translate English sentences into French. You want to give the model a few examples before asking it to translate a new sentence.

```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate

# Define examples of input and output (few-shot examples)
examples = [
    {"input": "What is your name?", "output": "Quel est votre nom ?"},
    {"input": "How are you?", "output": "Comment allez-vous ?"},
    {"input": "Where is the library?", "output": "Où est la bibliothèque ?"},
]

# Define the example template using PromptTemplate
example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="""
Input: {input}
Output: {output}
"""
)

# Create the FewShotPromptTemplate
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,  # Use the PromptTemplate object here
    prefix="Translate the following English sentences into French:",
    suffix="\nInput: {input}\nOutput:",
    input_variables=["input"],
)

# Now, let's format a new input into the few-shot prompt
formatted_prompt = few_shot_prompt.format(input="What time is it?")

print(formatted_prompt)

```

#What is an Agent?

In LangChain, an agent refers to a system where a language model can make decisions and take actions, such as calling APIs, retrieving information, or using tools. Unlike a static prompt where the model gives a direct answer, an agent dynamically selects which action to take based on the input and context.

#Agents typically:

Receive input (e.g., a question or task).
Decide on a course of action based on available tools (like calling APIs, searching the web, etc.).

Return results based on those actions.

#Components of an Agent:

Tools:

These are external functions or APIs the agent can call. For example, you might have tools for searching the web, performing calculations, or interacting with databases.

Agent:

The core logic that decides which tool to call and how to use it.

Agent Executor:

This is the piece that actually runs the agent, allowing it to use its tools and return the output.

#0ne confusion here agent will decide or llm will decide which tool to call ?

the LLM decides which tool to call based on the prompt and the input it receives. The agent is a framework that allows the LLM to access and use tools, but the decision-making comes from the LLM itself. Here’s how it works in more detail:

#How Agents Work:

Input is given to the agent: You provide the agent with an input (e.g., a question or task).

#LLM analyzes the input:

The language model (LLM) processes the input and interprets what the task is.
LLM decides which tool to use: The LLM, based on its understanding of the input and the tools available, decides if a tool needs to be used and which one. It might even decide not to use a tool at all if it can answer the question directly.

#Agent executes the tool:

Once the LLM chooses the tool, the agent framework executes the tool and processes its result.

#LLM integrates the result:

The LLM processes the tool's output and generates a response based on that result.

#Key Role of the LLM:

The LLM is essentially "thinking" and deciding which tool to use. The agent provides the interface and logic for executing the tools but does not itself decide when to use a tool.

#Example Breakdown:

Let's say the input is: "What is 12 plus 45 divided by 3?"

#LLM Processing:

The LLM reads the input and realizes it is a mathematical question.

#LLM Decides:

The LLM understands that it cannot do the calculation directly, so it decides to call the Calculator tool.

#Agent Executes:

The agent executes the Calculator tool, which evaluates 12 + 45 / 3 = 27.

#LLM Uses the Result:

The LLM takes the result of the tool (27) and formats the final response: "The result is 27."

# Agent Example 1:

Let’s look again at how this process happens with the LangChain agent:

```python
from langchain import OpenAI
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType

# Example tool (calculator)
def calculator(expression: str) -> str:
    """Evaluate a math expression."""
    try:
        return str(eval(expression))
    except Exception as e:
        return str(e)

# Initialize the OpenAI language model
llm = OpenAI(temperature=0)

# Define the tools inside tools we pass the Tool Object having function name , calculator and description .
tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Use this tool to calculate math expressions."
    ),
]

# Initialize the agent
agent = initialize_agent(
    tools=tools,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    llm=llm,
    verbose=True
)

# Run the agent with a math query
response = agent.run("What is 12 plus 45 divided by 3?")
print(response)
```

```python
import requests
from langchain.tools import tool
from langchain import OpenAI
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType

# Define the SerpApi key (replace 'your-serpapi-key' with your actual key)
# GET YOUR KEY FROM HERE :

# https://serpapi.com/manage-api-key

SERPAPI_KEY = ''

# Define the SerpApi search tool
@tool
def serpapi_search(query: str) -> str:
    """Search using SerpApi."""
    url = f"https://serpapi.com/search"
    params = {
        'api_key': SERPAPI_KEY,
        'q': query,
        'engine': 'google'  # Specify the search engine
    }
    response = requests.get(url, params=params)
    data = response.json()

    # Get the first result
    organic_results = data.get('organic_results', [])
    if organic_results:
        # Extract title and link of the first result
        first_result = organic_results[0]
        title = first_result.get('title', 'No title found')
        link = first_result.get('link', 'No link found')
        return f"Title: {title}\nLink: {link}"
    else:
        return "No results found."

# Initialize the OpenAI language model (you need an API key for this)
llm = OpenAI(temperature=0)

# Define the tools
tools = [
    Tool(
        name="Google Search",
        func=serpapi_search,
        description="Use this tool to search for information using Google."
    ),
]

# Initialize the agent with SerpApi as a tool
agent = initialize_agent(
    tools=tools,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    llm=llm,
    verbose=True
)

# Run the agent with a search query
response = agent.run("Find the latest information about LangChain framework.")
print(response)
```

# Chains In LangChain Concept

In LangChain, chains are sequences of steps or actions that can include interactions with language models, tools, or data sources. They allow you to structure and manage workflows by defining a series of operations that process input data and produce outputs.



# Types Of Chains

#Chain Types

##LLMChain:

Used for generating responses from a language model based on a prompt. It involves a single step where the language model generates output based on the input.

##SimpleSequentialChain:

Executes a sequence of operations where each step’s output is used as the input for the next step. Useful for tasks that involve multiple sequential steps.
ConversationChain: Manages ongoing conversations, maintaining context and dialogue history.

##MapReduceChain:

Applies a map-reduce approach, where a map function processes data in parallel, and a reduce function aggregates the results.

##SQLChain:

Interacts with SQL databases, allowing you to query and manipulate database records.

# Example 1 Simple LLM Chaining

###LLMChain

Description: A chain that focuses on generating outputs from a language model based on a given prompt.

```python
from langchain.chains import LLMChain
from langchain.llms import OpenAI

# Initialize the language model
llm = OpenAI(temperature=0)

# Define the LLMChain
chain = LLMChain(
    llm=llm,
    prompt_template="Translate the following text to French: {text}"
)

# Run the chain
response = chain.run({"text": "Hello, how are you?"})
print(response)  # Output: "Bonjour, comment ça va ?"
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import SequentialChain
from langchain.chains import LLMChain
from langchain.llms import OpenAI

# Initialize the OpenAI language model
client = OpenAI(temperature=0.7)

# Define the first prompt template to suggest a restaurant name
prompt_template_name = PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restaurant that serves {cuisine} cuisine. Suggest a fancy name for it."
)

# Chain 1: For generating the restaurant name
name_chain = LLMChain(
    llm=client,
    prompt=prompt_template_name,
    output_key="restaurant_name"  # Output key to pass to the next chain
)

# Define the second prompt template to suggest menu items based on the restaurant name
prompt_template_items = PromptTemplate(
    input_variables=["restaurant_name"],
    template="Suggest some menu items for the restaurant named {restaurant_name}."
)

# Chain 2: For generating menu items based on the restaurant name
food_items_chain = LLMChain(
    llm=client,
    prompt=prompt_template_items,
    output_key="menu_items"  # Key used to capture menu items as output
)

# Sequential chain to pass output from one chain to another
chain = SequentialChain(
    chains=[name_chain, food_items_chain],
    input_variables=["cuisine"],
    output_variables=["restaurant_name", "menu_items"],
    verbose=True
)

# Run the chain with the input as a dictionary
output = chain.run({"cuisine": "indian"})

print(output)

#Output
'''
{'cuisine': 'indian',
 'restaurant_name': '\n\n"The Maharaja\'s Palace"',
 'menu_items': '\n\n-Tandoori Chicken\n-Butter Chicken\n-Matar Paneer\n-Palak Paneer\n-Chana Masala\n-Aloo Gobi\n-Rogan Josh\n-Korma\n-Biryani\n-Naan\n-Raita\n-Kheer'}

'''
```

# Example Conversational Chain

###ConversationChain

###Description:

A chain designed for handling conversational contexts and maintaining dialogue history.

###How the Memory Works:

#####Initial Input:

When you ask the first question, "What is LangChain?", the model responds based on that input alone.

#####Subsequent Inputs:

When you ask the follow-up question, "Can you explain its features?", the model uses the stored memory to understand that "its" refers to "LangChain".

#####Updating Memory:

After each interaction, the memory is updated so the chain can keep track of the entire conversation. This ensures the model has context when you ask questions like "How can I use it in my project?"

```python
from langchain.chains import ConversationChain
from langchain.llms import OpenAI
from langchain.memory import ConversationMemory

# Initialize the language model
llm = OpenAI(temperature=0)

# Define the conversation chain
chain = ConversationChain(
    llm=llm,
    memory=ConversationMemory()
)

# Run the chain in a conversational context
response1 = chain.run("Hello, what is LangChain?")
print(response1)  # Output: "LangChain is a framework for building applications with language models."
response2 = chain.run("Can you tell me more about its features?")
print(response2)  # Output: "LangChain provides various features for integrating and using language models in applications."
```
