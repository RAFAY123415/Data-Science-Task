```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# please enter your file location in read_csv
gapminder=pd.read_csv(r'C:\Users\Zigron\Downloads\c2.csv')
```

```python
Pressure_and_U=gapminder[['hour.inc','U','pressure']]
```

```python
Pressure_and_U=Pressure_and_U.groupby('hour.inc').mean().reset_index()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
lns1=ax.plot(Pressure_and_U['hour.inc'], Pressure_and_U['U'], color="black", marker="o")
# set x-axis label
ax.set_xlabel("Hour.inc",color="blue",fontsize=15)
# set y-axis label
ax.set_ylabel("Average_U",color="red",fontsize=15)
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
lns2=ax2.plot(Pressure_and_U['hour.inc'], Pressure_and_U['pressure'],color="blue",marker="o")
ax2.set_ylabel("Average_Pressure",color="red",fontsize=15)
plt.gcf().set_size_inches(20, 5)

fig.legend(["Average_U","Average_Pressure"],loc="lower right", bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
plt.xlim(0,min(gapminder['hour.inc']))
plt.show()
```
