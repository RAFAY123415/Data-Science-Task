```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
flight_visuals=pd.read_csv("/content/flights_data.csv")
print(flight_visuals.shape)
```

```python
flight_visuals.head(2)
```

# BAR CHART

```python
sb.countplot(data=flight_visuals,x='Source')
plt.ylabel("Number of Flights",fontsize=12)
plt.xlabel('Source',fontsize=12)
plt.show()
```

```python
base_color=sb.color_palette()[0]
sb.countplot(data=flight_visuals,x='Source',color=base_color)
plt.xticks(rotation=30)
plt.show()
```

```python
base_color=sb.color_palette()[1]
gen_order=flight_visuals['Source'].value_counts().index
sb.countplot(data=flight_visuals,x='Source',color=base_color,order=gen_order)
plt.show()
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flight_visuals,x='Airline',color=base_color)
plt.show()
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flight_visuals,x='Airline',color=base_color)
plt.xticks(rotation=90)
plt.show()
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=flight_visuals,y='Airline',color=base_color)
```

# Data Missing Values

```python
flight_visuals.isna().sum()
```

```python
na_counts=flight_visuals.isna().sum()
base_color=sb.color_palette()[0]
sb.barplot(na_counts.index,na_counts,color=base_color)
plt.xticks(rotation=90)
plt.ylabel("Number of missing values",fontsize=12)
plt.show()
```

# Pie Chart

```python
sorted_counts=flight_visuals["Destination"].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index,startangle=90,counterclock=False)
plt.axis("square")
plt.title("Flight Destinations")
plt.show()
```

```python
sorted_counts=flight_visuals["Destination"].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index,startangle=90,counterclock=False,wedgeprops={'width':0.4})
plt.axis("square")
plt.show()
```
